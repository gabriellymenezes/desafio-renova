```python
# Imports
import pandas as pd
from IPython.display import display
```

```python
# Arquivo
df_sp = pd.read_csv('SP_turno_1.csv',sep=';',encoding='latin-1')
print(df_sp)
```

```python
# Remover colunas não necessarias
remover_colunas_sp = ['DT_GERACAO','HH_GERACAO','ANO_ELEICAO','CD_TIPO_ELEICAO','NM_TIPO_ELEICAO','CD_PLEITO','DT_PLEITO','NR_TURNO','CD_ELEICAO','DS_ELEICAO','SG_UF','CD_MUNICIPIO','CD_CARGO_PERGUNTA','DT_BU_RECEBIDO','CD_TIPO_URNA','DS_TIPO_URNA','CD_CARGA_1_URNA_EFETIVADA','CD_CARGA_2_URNA_EFETIVADA','CD_FLASHCARD_URNA_EFETIVADA','DT_CARGA_URNA_EFETIVADA','DS_CARGO_PERGUNTA_SECAO','CD_TIPO_VOTAVEL','DS_AGREGADAS','DT_ABERTURA','DT_ENCERRAMENTO','QT_ELEITORES_BIOMETRIA_NH','DT_EMISSAO_BU','NR_JUNTA_APURADORA','NR_TURMA_APURADORA','DS_TIPO_VOTAVEL', 'NR_URNA_EFETIVADA']
df_sp_cleaned = df_sp.drop(columns=remover_colunas_sp)

# Salvar os dados filtrados
df_sp_cleaned.to_csv('SP_turno_1.csv', sep=';',encoding='latin-1', index=False)
```
