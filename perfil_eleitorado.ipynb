```python
# Imports
import pandas as pd
from IPython.display import display
```

```python
# Arquivo
df_perfil = pd.read_csv('perfil_eleitorado_2020.csv',sep=';',encoding='latin-1')
print(df_perfil)
```

```python
# Remover as linhas com valores nulos da tabela especifica
df_perfil.dropna(subset=['QT_ELEITORES_PERFIL'], inplace=True)

# Transformar os dados da tabela em inteiros 
df_perfil['QT_ELEITORES_PERFIL'] = df_perfil['QT_ELEITORES_PERFIL'].astype(int)
```

```python
# Limpar brancos e nulos
df_perfil = df_perfil.dropna()

# Salvar os dados filtrados
df_perfil.to_csv('perfil_eleitorado_2020.csv', index=False)
```

```python
# Remover colunas não necessarias
remover_colunas = ['DT_GERACAO','ANO_ELEICAO', 'HH_GERACAO','CD_MUNICIPIO','CD_MUN_SIT_BIOMETRIA','DS_MUN_SIT_BIOMETRIA','CD_GENERO', 'CD_ESTADO_CIVIL','CD_FAIXA_ETARIA','CD_GRAU_ESCOLARIDADE', 'QT_ELEITORES_BIOMETRIA', 'QT_ELEITORES_DEFICIENCIA', 'QT_ELEITORES_INC_NM_SOCIAL']
df_perfil_cleaned = df_perfil.drop(columns=remover_colunas)

# Salvar os dados filtrados
df_perfil_cleaned.to_csv('perfil_eleitorado_2020.csv', sep=';',encoding='latin-1', index=False)
```

```python
# Exibir apenas DataFrame da UF SP
df_perfil_cleaned = df_perfil_cleaned[df_perfil_cleaned['SG_UF'] == 'SP']

# Salvar os dados filtrados
df_perfil_cleaned.to_csv('perfil_eleitorado_2020.csv', sep=';',encoding='latin-1', index=False)

# Exibir as primeiras linhas do DataFrame da UF SP
print(df_perfil_cleaned.head())
```

```python
# Remover a coluna SG_UF
remover_sg_uf = ['SG_UF']
df_perfil_cleaned = df_perfil_cleaned.drop(columns=remover_sg_uf)

# Salvar os dados filtrados
df_perfil_cleaned.to_csv('perfil_eleitorado_2020.csv', sep=';',encoding='latin-1' ,index=False)

# Exibir as primeiras linhas do DataFrame
print(df_perfil_cleaned.head())
```
